# problem statement
 will coustomer buy the product under following conditions which are:

    1. type of day -> weekday, hoilday, weekend

    2. avail of discount
 
    3. avail of free delivery
 
 target value ---> purchased or not purchased

```python
# from dataframe manipulation
import pandas as pd
```

```python
path = r"E:\coding\CSV\Naive_Bayes_Dataset.xlsx"
df = pd.read_excel(path)
df.shape
```

```python
# converting yes or no in binary format i.e 1 or 0
df.replace({"Yes":1, "No":0}, inplace=True)

# converting type of day in numerical value
df["Day"].replace({"Weekday":1, "Holiday":2, "Weekend":3}, inplace=True)

df
```

```python
# feature are independent value
feature = df[["Day","Discount","Free Delivery"]]

# target are the dependent value
target = df[["Purchase"]]
```

```python
feature
```

```python
target
```

```python
# spliting data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.20, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# using naive bayes model for prediction
from  sklearn.naive_bayes import GaussianNB

# GaussianNB model for prediction is used 
nb_model = GaussianNB()
```

```python
nb_model = nb_model.fit(X_train,y_train)
nb_model
```

```python
nb_model.score(X_test,y_test)
```

```python
nb_model.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()

DTC = DTC.fit(X_train,y_train)
DTC
```

```python
DTC.score(X_test,y_test)
```
